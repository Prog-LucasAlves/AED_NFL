### Bibliotecas Utilizadas:

```python
# Manipulação de dados
import pandas as pd
import numpy as np

# Visualização - Gráficos
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns
import hvplot.pandas 

# Ignore Warning(Avisos)
import warnings

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', 100)
```

### Carregando os dados

```python
data = pd.read_csv('../../DATA/DATA_FOR_CLEANING/PASSING/PASSING_CLEAR.csv')
```

### Criando Filtro por Jogador

```python
nome_jogardor = data[data['Year'] == 2022].sort_values(by='Player')
print(list(nome_jogardor['Player']))
```

```python
# Criando o filtro com o nome do jogador escolhido - Digite um dos nome acima
jogador = 'Tom Brady'

# Filtrando na base de dados pelo jogador escolhido e salvando em um novo DataFrame
data_escolha = data[data['Player'] == jogador]
```

### Gráfico

```python
indice = np.arange(len(data['Year']))

plt.style.use('seaborn-pastel')
sns.set(font_scale=1)

figura1, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,3))
figura2, (ax3, ax4) = plt.subplots(1, 2, figsize=(15,3))
figura3, (ax5, ax6) = plt.subplots(1, 2, figsize=(15,3))
figura4, (ax7, ax8) = plt.subplots(1, 2, figsize=(15,3))
figura5, (ax9, ax10) = plt.subplots(1, 2, figsize=(15,3))
figura6, (ax11, ax12) = plt.subplots(1, 2, figsize=(15,3))
figura7, (ax13, ax14) = plt.subplots(1, 2, figsize=(15,3))

# Primeiro bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['Att'], ax=ax1, data=data_escolha)
ax1.axhline(data_escolha['Att'].mean(), label=f"Média - {round(data_escolha['Att'].mean(),2)}", linewidth=1, color='#d62728')
ax1.legend()
ax1.set_title(f'Total de Tentativas de Passe * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data['Cmp'], ax=ax2, data=data_escolha)
ax2.axhline(data_escolha['Cmp'].mean(), label=f"Média - {round(data_escolha['Cmp'].mean(),2)}", linewidth=1, color='#d62728')
ax2.legend()
ax2.set_title(f'Total de Passes Completos * Ano * {jogador}')

# Segundo bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['Cmp%'], ax=ax3, data=data_escolha)
ax3.axhline(data_escolha['Cmp%'].mean(), label=f"Média - {round(data_escolha['Cmp%'].mean(),2)}", linewidth=1, color='#d62728')
ax3.legend()
ax3.set_title(f'Porcentagem de Passes Completos * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data_escolha['Yds'], ax=ax4, data=data_escolha)
ax4.axhline(data_escolha['Yds'].mean(), label=f"Média - {round(data_escolha['Yds'].mean(),)}", linewidth=1, color='#d62728')
ax4.legend()
ax4.set_title(f'Total de Jardas - Passe * Ano * {jogador}')

# Terceiro bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['TD'], ax=ax5, data=data_escolha)
ax5.axhline(data_escolha['TD'].mean(), label=f"Média - {round(data_escolha['TD'].mean(),)}", linewidth=1, color='#d62728')
ax5.legend()
ax5.set_title(f'Passes para Touchdowns * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data_escolha['TD/G'], ax=ax6, data=data_escolha)
ax6.axhline(data_escolha['TD/G'].mean(), label=f"Média - {round(data_escolha['TD/G'].mean(),2)}", linewidth=1, color='#d62728')
ax6.legend()
ax6.set_title(f'Média de Touchdowns por jogo * Ano * {jogador}')

# Quarto bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['Int'], ax=ax7, data=data_escolha)
ax7.axhline(data_escolha['Int'].mean(), label=f"Média - {round(data_escolha['Int'].mean(),2)}", linewidth=1, color='#d62728')
ax7.legend()
ax7.set_title(f'Interceptações de Passes * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data_escolha['Int/G'], ax=ax8, data=data_escolha)
ax8.axhline(data_escolha['Int/G'].mean(), label=f"Média - {round(data_escolha['Int/G'].mean(),2)}", linewidth=1, color='#d62728')
ax8.legend()
ax8.set_title(f'Média de Interceptações de Passes por jogo * Ano * {jogador}')

# Quinto bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['Y/G'], ax=ax9, data=data_escolha)
ax9.axhline(data_escolha['Y/G'].mean(), label=f"Média - {round(data_escolha['Y/G'].mean(),2)}", linewidth=1, color='#d62728')
ax9.legend()
ax9.set_title(f'Média de Jardas por Jogo * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data_escolha['Y/A'], ax=ax10, data=data_escolha)
ax10.axhline(data_escolha['Y/A'].mean(), label=f"Média - {round(data_escolha['Y/A'].mean(),2)}", linewidth=1, color='#d62728')
ax10.legend()
ax10.set_title(f'Média de Jardas Ganhas por Tentativa de Passe * Ano * {jogador}')

# Sexto bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['Sk'], ax=ax11, data=data_escolha)
ax11.axhline(data_escolha['Sk'].mean(), label=f"Média - {round(data_escolha['Sk'].mean(),2)}", linewidth=1, color='#d62728')
ax11.legend()
ax11.set_title(f'Quantidade de Sacks * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data_escolha['Sk/G'], ax=ax12, data=data_escolha)
ax12.axhline(data_escolha['Sk/G'].mean(), label=f"Média - {round(data_escolha['Sk/G'].mean(),2)}", linewidth=1, color='#d62728')
ax12.legend()
ax12.set_title(f'Média de Sacks por Jogo * Ano * {jogador}')

# Setimo bloco de gráficos
sns.barplot(x=data_escolha['Year'], y=data_escolha['SYds'], ax=ax13, data=data_escolha)
ax13.axhline(data_escolha['SYds'].mean(), label=f"Média - {round(data_escolha['SYds'].mean(),2)}", linewidth=1, color='#d62728')
ax13.legend()
ax13.set_title(f'Jardas Perdidas por Sacks * Ano * {jogador}')

sns.barplot(x=data_escolha['Year'], y=data_escolha['SYds/G'], ax=ax14, data=data_escolha)
ax14.axhline(data_escolha['SYds/G'].mean(), label=f"Média - {round(data_escolha['SYds/G'].mean(),2)}", linewidth=1, color='#d62728')
ax14.legend()
ax14.set_title(f'Média de Jardas Perdidas por Sack por Jogo * Ano * {jogador}');
```

### Temporada 22/23

```python
# Criando o filtro temporada
temp = 2022

# Filtrando na base de dados pela temporada e salvando em um novo DataFrame
data_tem = data[(data['Year'] == temp) & (data['Cmp'] >= 100)]

# Visualizando
data_tem.tail()
```

```python
# Selecionando Variáveis
# Top 5 jardas
data_yds = data_tem[['Player', 'Yds']].sort_values(by='Yds', ascending=False).head(5)

# Top 5 jardas por jogo
data_game = data_tem[['Player', 'Y/G']].sort_values(by='Y/G', ascending=False).head(5)

# Top 5 Passe completo mais Longo
data_lng = data_tem[['Player', 'Lng']].sort_values(by='Lng', ascending=False).head(5)

# Top 5 passes completos
#data_att = data_tem[['']]

# Top 5 Passes completos %
data_cmp = data_tem[['Player', 'Att', 'Cmp', 'Cmp%']].sort_values(by='Cmp%', ascending=False).head(5)

# Top 5 Interceptações 
data_int = data_tem[['Player', 'Int']].sort_values(by='Int', ascending=False).head(5)

# Configuração dos gráficos
figura1, ((ax100, ax101, ax102), (ax103, ax104, ax105)) = plt.subplots(2, 3, figsize=(15,8))
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)

# Primeiro bloco de gráficos
#1.1
sns.barplot(x=data_yds['Yds'], y=data_yds['Player'], ax=ax100, data=data_yds)
ax100.bar_label(ax100.containers[0], fmt=f'%.0f Yds', padding=-54)
ax100.set_title(f'Top 5 Total de Jardas Temp. 22/23')
ax100.set_xlim(right=2800)

#1.2
sns.barplot(x=data_game['Y/G'], y=data_game['Player'], ax=ax101, data=data_game)
ax101.bar_label(ax101.containers[0], fmt=f'%.0f Y/G', padding=-46)
ax101.set_title(f'Top 5 Jardas por Jogo Temp. 22/23')
ax101.set_xlim(right=350)

#1.3
sns.barplot(x=data_lng['Lng'], y=data_lng['Player'], ax=ax102, data=data_lng)
ax102.bar_label(ax102.containers[0], fmt=f'%.0f Yds', padding=-35)
ax102.set_title(f'Top 5 O Passe Completo Mais Longo Lançado Temp. 22/23')
ax102.set_xlim(right=100)

"""
#1.2
sns.barplot(x=data_tds['TD'], y=data_tds['Player'], ax=ax101, data=data_tds)
ax101.bar_label(ax101.containers[0], fmt=f'%.0f TDs', padding=-42)
ax101.set_title(f'Top 5 Total de Touchdowns Temp. 22/23')
ax101.set_xlim(left=0, right=25)
"""

#2.1

"""
sns.barplot(x=data_cmp['Cmp%'], y=data_cmp['Player'], ax=ax102, data=data_cmp)
ax102.bar_label(ax102.containers[0], fmt=f'%.2f%%', padding=-48)
#labels=[f'%.2f' % e for e in data_cmp['Cmp%']]
#labels=[f"{height:.0%}"]
ax102.set_title(f'\nTop 5 Porcentagem de Passes Completos (Acima de 100 passes) Temp. 22/23')
ax102.set_xlim(right=85)

#2.2
sns.barplot(x=data_int['Int'], y=data_int['Player'], ax=ax103, data=data_int)
ax103.bar_label(ax103.containers[0], fmt=f'%.0f Int', padding=-30)
ax103.set_title(f'\nTop 5 Total Interceptações Temp. 22/23')
ax103.set_xlim(left=0, right=10)
"""

figura1.tight_layout() 
```

```python
# Tentativa de Passes x Passes completos x Touchdowns Acima de 100 Passes - Temp 22/23

fig = go.Figure()
fig.add_trace(go.Bar(
    y=data_tem['Player'],
    x=data_tem['Att'],
    name='Tentativa de Passes',
    orientation='h',
    
))
fig.add_trace(go.Bar(
    y=data_tem['Player'],
    x=data_tem['Cmp'],
    name='Passes Completos',
    orientation='h',
    
))
fig.add_trace(go.Bar(
    y=data_tem['Player'],
    x=data_tem['TD'],
    name='Touchdowns',
    orientation='h',
    
))
fig.update_layout(title='Tentativa de Passes x Passes completos x Touchdowns <br> Acima de 100 Passes - Temp 22/23', barmode='stack', height=700, width=700)
fig.update_yaxes(title='Player')
fig.update_xaxes(title=' Passes - Touchdowns ')
fig.show()
```
